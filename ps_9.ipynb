<h3>1. Create Data Subsets for Type of Post</h3>

<h4>Getting unique post types</h4>

```python
df['Type'].unique()
```

<h4>Getting size of each dataset</h4>

```python
no_of_photos = (df['Type'] == "Photo").sum()
no_of_status = (df['Type'] == "Status").sum()
no_of_links = (df['Type'] == "Link").sum()
no_of_videos = (df['Type'] == "Video").sum()

print(f"No. of Photo Posts: {no_of_photos}")
print(f"No. of Status Posts: {no_of_status}")
print(f"No. of Link Posts: {no_of_links}")
print(f"No. of Video Posts: {no_of_videos}")
```

<h4>Creating seperate datasets for each type</h4>

```python
photo_df = df[df['Type'] == "Photo"]
status_df = df[df['Type'] == "Status"]
link_df = df[df['Type'] == "Link"]
video_df = df[df['Type'] == "Video"]
```

<h3>2. Merge Two Subsets (photo and video)</h3>

```python
multimedia_df = pd.concat([photo_df, video_df])
```
