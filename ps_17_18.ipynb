```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
cols = ['age', 'sex', 'cp', 'trestbps', 'chol', 'fbs', 'restecg', 'thalach', 'exang', 'oldpeak', 'slpoe', 'ca', 'thal', 'num']
df = pd.read_csv("datasets/HeartDisease/Cleavland.csv", names=cols, header=None)
```

<h3>1. Histogram</h3>

```python
df[cols].hist(bins=10, figsize=(10,10))
```

<h3>2. Dot Plot</h3>

<h3>3. Bar Plot</h3>

```python
for col in cols:
    df[col].value_counts().plot(kind='bar')
    plt.xlabel(col)
    plt.ylabel('Count')
    plt.show()
```

<h3>4. Line Chart</h3>

```python
for col in cols:
    plt.plot(df[col])
    plt.xlabel('Index')
    plt.ylabel(col)
    plt.show()
```

<h3>5. Scatter Plot</h3>

```python
cols = ['chol', 'thalach']
sns.scatterplot(data=df, x=cols[0], y=cols[1], hue='num')
plt.xlabel(cols[0])
plt.ylabel(cols[1])
plt.show()
```
