<h3>1. Create data subsets for different languages</h3>

```python
languages = df['original_language'].unique()
languages
```

```python
language_dfs = []

for language in languages:
    language_dfs.append(df[df['original_language'] == language])
```

```python
language_fds_dict = dict(zip(languages, language_dfs))
```

```python
language_fds_dict
```

<h3>2. Merge two subsets</h3>

```python
merged_df = pd.concat([language_fds_dict['en'], language_fds_dict['fr']])
```

```python
merged_df
```

<h3>3. Sort Data using customer ratings.</h3>

```python
sorted_df = df.sort_values(by='vote_average', ascending=False)
```

```python
sorted_df
```

<h3>4. Transposing Data</h3>

```python
transposed_df = df.transpose()
```

<h3>5. Melting Data to long format</h3>

```python
melted_df = pd.melt(df, id_vars=['id','original_language'])
```

```python
melted_df
```

<h3>6. Casting data to wide format</h3>

```python
pivoted_table = pd.pivot_table(df, index=['id','original_language'], values=['original_title','spoken_languages'])
```
